```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from facenet_pytorch import MTCNN
import os, cv2
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu") # device object
device
```

```python
mtcnn = MTCNN(keep_all=True, device=device)
new_img_dir = './input/data/train/cropped_images'
img_path = './input/data/train/images'

cnt = 0

for paths in os.listdir(img_path):
    if paths[0] == '.': continue
    
    sub_dir = os.path.join(img_path, paths)
    
    for imgs in os.listdir(sub_dir):
        if imgs[0] == '.': continue
        
        img_dir = os.path.join(sub_dir, imgs)
        img = cv2.imread(img_dir)
        img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        
        #mtcnn 
        boxes,probs = mtcnn.detect(img)
        
        # boxes 
        if len(probs) > 1: 
            print(boxes)
        if not isinstance(boxes, np.ndarray):
            print('Nope!')
            #  crop
            img=img[100:400, 50:350, :]
        
        # boexes size 
        else:
            xmin = int(boxes[0, 0])-30
            ymin = int(boxes[0, 1])-30
            xmax = int(boxes[0, 2])+30
            ymax = int(boxes[0, 3])+30
            
            if xmin < 0: xmin = 0
            if ymin < 0: ymin = 0
            if xmax > 384: xmax = 384
            if ymax > 512: ymax = 512
            
            img = img[ymin:ymax, xmin:xmax, :]
        
        n_path = os.path.join(paths, imgs).replace('/', '_')
        cnt += 1
        print(os.path.join(new_img_dir, n_path))
        plt.imsave(os.path.join(new_img_dir, n_path), img)
        
print(cnt)
```

```python
mtcnn = MTCNN(keep_all=True, device=device)
new_img_dir = './input/data/eval/cropped_images'
img_path = './input/data/eval/images'

cnt = 0
    
for imgs in os.listdir(img_path):
    if imgs[0] == '.': continue

    img_dir = os.path.join(img_path, imgs)
    img = cv2.imread(img_dir)
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

    #mtcnn 
    boxes,probs = mtcnn.detect(img)

    # boxes 
    if len(probs) > 1: 
        print(boxes)
    if not isinstance(boxes, np.ndarray):
        print('Nope!')
        #  crop
        img=img[100:400, 50:350, :]

    # boexes size 
    else:
        xmin = int(boxes[0, 0])-30
        ymin = int(boxes[0, 1])-30
        xmax = int(boxes[0, 2])+30
        ymax = int(boxes[0, 3])+30

        if xmin < 0: xmin = 0
        if ymin < 0: ymin = 0
        if xmax > 384: xmax = 384
        if ymax > 512: ymax = 512

        img = img[ymin:ymax, xmin:xmax, :]

    cnt += 1
    plt.imsave(os.path.join(new_img_dir, imgs), img)
        
print(cnt)
```

# Save cropped dataframe as csv file

```python
data_dir = 'input/data/train'
df_path = f'{data_dir}/new_train.csv'
cropped_df_path = f'{data_dir}/cropped_train.csv'
```

```python
df = pd.read_csv(df_path)
df
```

```python
for i in range(len(df)):
    path = df.iloc[i]["abs_path"].replace('images', 'cropped_images')
    tmp_path = path[32:].replace('/', '_')
    path = path[:32] + tmp_path
    df.loc[i, "abs_path"] = path

df.head()
```

```python
df.to_csv(cropped_df_path)
```
